## Check training data size dependency for RBM reconstruction

- Trained respective RBM using `range(500, 20000, 1000)` data and see
 - Data size dependency in each epoch
 - as well as epoch-dependency difference in each training data size

### Learning curve w.r.t. training data size
- Lines for each training data size

```python
import numpy as np
import pandas as pd

df = pd.read_csv('training_logs/task2_2_train_num.csv', names=list(range(100, 1001, 100)))
df.index = list(range(500, 20000, 1000))
ax = df.T.plot()
ax.set_yscale('log')
ax.set_ylabel('|ΔE|')
ax.set_xlabel('Epoch')
```

### Training data size dependency for each epoch checkpoints
- Lines for each epoch
- Log-log scale
- We can clearly see that increasing training data size makes learning curve steeper and also makes learning efficient.
- Shows power law

```python
ax = df.plot()
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('|ΔE|')
ax.set_xlabel('Training data size')
```
